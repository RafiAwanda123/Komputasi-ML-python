```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
from  matplotlib.animation import FuncAnimation
```

Generate

```python
Jumlah_data = 100

y = np.array([i*0.1+np.random.randn() for i in range(Jumlah_data)])
x = np.array([i*0.1 for i in range(Jumlah_data)])
```

Plot Data


```python
plt.scatter(x, y)
```

Fungsi Linear


```python
def Sinau_linear(x, gradien):
    y = gradien * x
    return y
```

Menghitung linear awal

```python
x_prediksi = np.array([0, 10])
m_awal = 5
y_prediksi = Sinau_linear(x_prediksi, m_awal)
```

```python

#plt.scatter(x, y)
#plt.plot(x_prediksi, y_prediksi, "yellow")
#plt.axis([0,10,0,10])
#plt.show()
```

```python
m_prediksi = m_awal
m_list_prediksi = []
y_list_prediksi = []
x_list_prediksi = []
learning_rate = 0.1
for i in range(1, Jumlah_data):
    y_prediksi = Sinau_linear(x[i], m_prediksi)
    y_asli = y[i]
    
    error = y_asli-y_prediksi
    
    delta_m = learning_rate * error/x[i]
    
    m_prediksi = m_prediksi + delta_m
    
    m_list_prediksi.append(m_prediksi)
    y_list_prediksi.append(Sinau_linear(np.array([0, 10]), m_prediksi))
    x_list_prediksi.append(np.array([0, 10]))
```

Visualisai Prediksi

```python
fig = plt.figure()
line, = plt.plot([], [], "Yellow")

def animate(frame_num):
    x = x_list_prediksi[frame_num]
    y = y_list_prediksi[frame_num]
    line.set_data(x,y)
    return line

plt.scatter(x,y)
plt.axis([0, 10, 0, 10])
Ahhh = FuncAnimation(fig,animate,frames = 120, interval = 100, repeat = False)
plt.show()
```
