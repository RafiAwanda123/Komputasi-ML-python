# Komputasi Termodinamika 

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
RO = 3000
TO = 298
R = 200
T = 100 + 273.15
R = float(input("Masukin resistansi yang kamu mau(dalam ohm):"))
a = 1/(1/T-1/TO)*np.log(R/RO)
Buat_di_print = "Temperaturnya adalah {a} derajat celcius".format(a = a)
print(Buat_di_print)
```

```python
fig, ax = plt.subplots(1, figsize=(10,5))

Ts = np.linspace(0, 500)
Rs = RO*np.exp(a * (1/(Ts + 273.15) - 1/TO))

ax.plot(Ts, Rs)
ax.grid(True, alpha = 0.33)
ax.set_xlabel("T/oC", fontsize = 16)
ax.set_ylabel("R/Ohm", fontsize = 16)
print(ax)
```
