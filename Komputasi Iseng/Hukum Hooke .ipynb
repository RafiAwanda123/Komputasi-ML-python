# Gerak Sistem Pegas Bermassa 

Soal


Sebuah benda bermassa 7 kg yang dihubungkan dengan sistem pegas vertikal dengan konstanta pegas 1,8 N/m. Mula - Mula benda ditarik sejauh 8 cm kemudian dilepaskan sehingga benda berosilasi. Jika diasumsikan sistem pegas mendapat gaya hambatan udara dengan koefisien gaya hambatan 0,5 kg/s dan dengan menggunakan metode fenyman newton, buatlah program menggunakan python untuk menghitung dan memvisualisasikan posisi, kecepatan dan percepatan benda sebagai fungsi dar waktu, dan kemudian analisa bagaimana periode dari osilasi tersebut.


#### masukin libarinya dulu ya ges ya

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```



#### Masukin yang nilai yang diketahui

```python
k = 1.8 # N/m
m = 7 # kg
c = 0.5 # kg/s
y = -0.8e-2 # m
v = 0 # m/s
h = 0.1 # s
t = 0 
tmaks = 50 #s 
```




#### langsung kerumus 


```python
vFN = v
yFN = y 
Amplitudo = y
tt = [] #menyimpan data waktu
yyFN = [] # menyimpan data fymen
vvFN = [] # menyimpan kecepatan 
aaFN = [] # menyimpan Percepatan 
xmaksTT = [] # menyimpan data amplitudo maksimal
```

```python
while t < tmaks :
    
    #masukin rumus tadi yang udah dicari 
    aFN = -(k*yFN + c*vFN)/m
    xmaksT = Amplitudo * np.exp(-c*t / (2+m))
    tt.append(t)
    yyFN.append(yFN)
    vvFN.append(vFN)
    aaFN.append(aFN)
    xmaksTT.append(xmaksT)
    
    #setengah langkah
    ySet = yFN + vFN + h/2
    vSet = vFN + aFN + h/2
    aSet = -(k*ySet + c*vSet)/m
    yFN += h*vSet
    vFN += h*aSet
    t += h

aFN = -(k*yFN + c*vFN)/m
print('cuman buat ngeluarin hasil aja yang atas buat grafiknya ')
print(aFN)
```



#### langsung gaskan buat grafik 

```python
# buat background
fig, ax = plt.subplots(figsize=(15, 10), dpi = 100)

# buat plotnya
ax.plot(tt, vvFN, color='blue', marker = "", label = 'Kecepatan Feynman Newton')

# garis X
ax.set_xlabel('Waktu (s)', fontsize = 14)

# garis y 
ax.set_ylabel('Kecepatan (m/s)',color = 'black', fontsize = 14)
ax.legend(loc = 'upper right')

# membuat dua objek garis yang beda 
ax2 = ax.twinx()

# membuat grafik dengan beda garis y dengan axis objek
ax2.plot(tt, yyFN, color='red',  marker = "", label = 'Posisi Feynman Newton')
ax2.plot(tt, xmaksTT, '--', label = r'$y_{maks}(t)=Ae^{-ct/2m}$')
ax2.set_ylabel('Posisi (m) ', color = 'black', fontsize = 14)
ax2.legend(loc = 'upper left')
```
